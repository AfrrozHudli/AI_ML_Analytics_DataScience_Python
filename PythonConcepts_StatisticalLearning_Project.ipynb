### Question 7

Lets define **Goal difference** as the difference between the **GoalsFor** and **GoalsAgainst**. And it can take negative values also.

Print all the goal differences for all the teams.

Write a function with name **goal_diff_count** which should return all the teams with their Goal Differences.

Write a function with name **max_goal_diff_team** which should return the team with max Goal Difference.

Write a function with name **min_goal_diff_team** which should return the team with min Goal Difference.



```python
#Note functions have been written in the order in which they have been asked in the question.

laliga['GoalsFor']= laliga['GoalsFor'].astype('int64')
laliga['GoalsAgainst']= laliga['GoalsAgainst'].astype('int64')
#goal_diff_count():
def goal_diff_count(x,y):
    return x-y
goal_diff_count(laliga.GoalsFor, laliga.GoalsAgainst)
```

```python
#max_goal_diff_team():
def max_goal_diff_team():
    laliga['GoalDiffCount'] = goal_diff_count(laliga.GoalsFor, laliga.GoalsAgainst)
    return laliga[laliga['GoalDiffCount']==max(laliga.GoalDiffCount)]
max_goal_diff_team()
```

```python
#min_goal_diff_team():
def min_goal_diff_team():
    laliga['GoalDiffCount'] = goal_diff_count(laliga.GoalsFor, laliga.GoalsAgainst)
    return laliga[laliga['GoalDiffCount']==min(laliga.GoalDiffCount)]
min_goal_diff_team()
```

# Analysis of Donald Trump's Tweets 

### Question 1

## Read the dataset into a dataframe

Read the dataset **'trump_twitter.csv'** (taken from [FiveThirtyEight](https://data.fivethirtyeight.com/)) into a dataframe with name **trump_df** using pandas.

```python
import pandas as pd
import numpy as np
```

```python
trump_df = pd.read_csv('trump_twitter.csv')
trump_df
```

### Question 2

## Trump is really excited for running as President. 

Find the percentage of tweets with **atleast one '!'**.

```python
totalNoOfTweets=len(trump_df['text'])
n=0
totalNoOfAtLeastOneExclamation = 0
while n < len(trump_df['text']):
    trump_df.loc[n,'noOfExclaimations'] = trump_df.loc[n,'text'].count('!')
    if (trump_df.loc[n,'noOfExclaimations'] > 0):
        totalNoOfAtLeastOneExclamation = totalNoOfAtLeastOneExclamation + 1
    n=n+1
perct_of_tweets_with_atleast_one_exclamation = (totalNoOfAtLeastOneExclamation/totalNoOfTweets)*100
print('The percentage of tweets with atleast one "!" is ' + str(perct_of_tweets_with_atleast_one_exclamation))
```

### Question 3

## Number of Tweets Donald Trump posted per day

Plot the number of tweets per day trump posted.

Refer, if you are looking for Date splitting : http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.str.split.html


```python
#There can be two plots for this question 
#1. the dates left unsorted and
#2. the dates sorted to go from 2015 into 2016.
#I have plotted both, though the sorted plot is more insightful as we can comment on Trumps tweeting trend from it.

# Plot showing Tweet Dates on x-Axis as provided in the raw data feed i.e. unsorted

trump_df[['splitDate','splitTime']] = trump_df['created_at'].str.split(expand=True)
#trump_df
#trump_df.loc['splitDate']
by_day = trump_df.groupby('splitDate')
df = pd.DataFrame(by_day.count())
xAxis=df.index


#Code for plot
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
N=183
ind = np.arange(N)
fig, ax = plt.subplots(figsize=(20,7))
trump_df.groupby(['splitDate']).count()['splitTime'].plot(ax=ax)

ax.set_xlabel('Tweet Dates (unsorted)', fontsize=20)
ax.format_xdata = mdates.DateFormatter('%Y-%m-%d')
plt.xticks(ind, xAxis, fontsize=9)
plt.xticks(rotation=90)
ax.grid(True)

#plt.AutoLocator()
ax.set_ylabel('Number of Tweets ---->',fontsize=20)
ax.set_title('The number of tweets per day Donald Trump posted',fontsize=20)

#Zoom-in
#plt.xlim(1.3, 50) #Use this to zoom-in and check points on x-axis the second variable currently set to 183 
#change it to a number of points in the range of 1 to 183.

#Code for showing plot

plt.show()

#Since this plot has unsorted x-axis it's difficult to make any meaningful inference about Trump's tweeting trend.
```

```python
# Plot with ordered date on x Axis starting with 2015 going to 2016 i.e. x-Axis sorted

#saving the trump_df dataframe as trump_df2 dataframe so that orginal trump_df is not overwritten
trump_df2 = trump_df
#trump_df2.info()

#Now to split column created_at into date and time
trump_df2[['splitDate','splitTime']] = trump_df2['created_at'].str.split(expand=True)

#Now convert 'splitDate' to date time format and sort
trump_df2['splitDate'] = pd.to_datetime(trump_df2['splitDate'])
trump_df2=trump_df2.sort_values(by=['splitDate'])

#Grouping by day
by_day = trump_df2.groupby('splitDate')
#by_day.count()

#Count of tweets per day
noTweets = pd.DataFrame(by_day.count())
#noTweets
xAxis=noTweets.index
#xAxis

#Code for plot
import matplotlib.pyplot as plt2
%matplotlib inline
plt2.rcParams["figure.figsize"] = (20,7)
fig = plt2.figure()


# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 1, .8]) # left, bottom, width, height (range 0 to 1)
axes.grid(True)
plt2.xlabel('Date Tweeted on', fontsize = 18)
plt2.ylabel('Number of Tweets', fontsize = 18)
plt2.title('The number of tweets per day Donald Trump posted',fontsize = 18)
plt2.plot(xAxis,noTweets, 'g.-') # 'b' is the color blue

#Code for showing plot

plt2.show()

#Because this plot is with sorted X-Axis we can infer by looking at it that Trump' tweets declined post March 2016.
```
